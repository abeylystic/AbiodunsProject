```python
import pandas as pd
import random
nodes = ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J"]
edges = {source: {target:"" for target in nodes if target != source} for source in nodes}
state = ["K-Folds", "F-OLS", "Both"]

for source in edges.keys():
    for target in edges[source].keys():
        edges[source][target] = random.choice(state)

pd.DataFrame(edges).sort_index()
```

```python
k_folds = []
f_ols = []
for source in edges.keys():
    for target in edges[source].keys():
        if edges[source][target] == "Both": 
            k_folds.append((source, target))
            f_ols.append((source, target))
        elif edges[source][target] == "K-Folds":
            k_folds.append((source, target))
        elif edges[source][target] == "F-OLS":
            f_ols.append((source, target))
print("k-folds", k_folds)
print("f-ols", f_ols)
```

```python
network = {source: {target: "" for target in nodes if target != source} for source in nodes}
for source in edges.keys():
    for target in edges[source].keys():
        if (source, target) in k_folds and (source, target) in f_ols:
            network[source][target] = "Both"
        elif (source, target) in k_folds:
            network[source][target] = "K-Folds"
        elif (source, target) in f_ols:
            network[source][target] = "F-OLS"
        else:
            network[source][target] = "NaN"
network = pd.DataFrame(network).sort_index()
# calcluate percentage of links that are in both graphs
network.apply(lambda x: len(x[x == "Both"]), axis=1).sum() / (network.shape[0] * (network.shape[0] - 1))
```
