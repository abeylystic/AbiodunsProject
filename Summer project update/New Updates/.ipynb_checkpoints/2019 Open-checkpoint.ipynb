### Problem 1(a)

**Objective:** Assess whether the weight status of boys significantly differs from girls using BMI categories.

**Given Data:**

- **Boys:**
  - Underweight: 7
  - Normal weight: 115
  - Overweight: 21

- **Girls:**
  - Underweight: 17
  - Normal weight: 113
  - Overweight: 13

**Formula:**

To compare the weight status between boys and girls, we use the Chi-Square test of independence.

The Chi-Square statistic is calculated using:

$\chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}$

where $O_i$ is the observed frequency and $E_i$ is the expected frequency under the null hypothesis.

**Calculation:**

1. **Construct the contingency table:**

   |         | Underweight | Normal weight | Overweight | Total |
   |---------|--------------|---------------|------------|-------|
   | Boys    | 7            | 115           | 21         | 143   |
   | Girls   | 17           | 113           | 13         | 143   |
   | **Total** | **24**         | **228**          | **34**        | **286**   |

2. **Calculate the expected frequencies:**

   
   $E_{ij} = \frac{(Row\ Total_i) \times (Column\ Total_j)}{Grand\ Total}$
   

   - Expected frequency for boys underweight:
     
     $E_{1,1} = \frac{143 \times 24}{286} = 12$
     

   - Expected frequency for boys normal weight:
     
     $E_{1,2} = \frac{143 \times 228}{286} = 114$
     

   - Expected frequency for boys overweight:
     
     $E_{1,3} = \frac{143 \times 34}{286} = 17$
     

   - Expected frequency for girls underweight:
     
     $E_{2,1} = \frac{143 \times 24}{286} = 12$
     

   - Expected frequency for girls normal weight:
     
     $E_{2,2} = \frac{143 \times 228}{286} = 114$
     

   - Expected frequency for girls overweight:
     
     $E_{2,3} = \frac{143 \times 34}{286} = 17$
     

3. **Calculate the Chi-Square statistic:**

   
   $\chi^2 = \frac{(7-12)^2}{12} + \frac{(115-114)^2}{114} + \frac{(21-17)^2}{17} + \frac{(17-12)^2}{12} + \frac{(113-114)^2}{114} + \frac{(13-17)^2}{17}
   $

   
   $\chi^2 = \frac{25}{12} + \frac{1}{114} + \frac{16}{17} + \frac{25}{12} + \frac{1}{114} + \frac{16}{17}
   $

   $
   \chi^2 = 2.083 + 0.0088 + 0.941 + 2.083 + 0.0088 + 0.941
   $

   $
   \chi^2 = 6.0656
   $

4. **Determine the degrees of freedom:**

   $
   df = (number\ of\ rows - 1) \times (number\ of\ columns - 1) = (2 - 1) \times (3 - 1) = 2
   $

5. **Compare the calculated Chi-Square statistic with the critical value:**

   Using a Chi-Square distribution table, for df = 2 and $\alpha$ = 0.05, the critical value is 5.991.

   Since $\chi^2_{calculated}$ = 6.0656 is greater than $\chi^2_{critical}$ = 5.991, we reject the null hypothesis.

**Interpretation:** There is a significant difference in the weight status distribution between boys and girls at the 5% significance level.

### Problem 1(b)

**Objective:** Compare the distribution of serum estradiol for Caucasian vs. African American women.

**Given Data:**

- **Caucasian:** 14, 25, 28, 31, 32, 41, 56, 65, 71, 94
- **African American:** 12, 18, 19, 21, 22, 35, 36, 39, 42, 43, 44, 46, 50, 54, 91

**Formula:**

To compare the distributions, we can use a two-sample t-test assuming unequal variances (Welch's t-test).

The t-test statistic is calculated using:
$
t = \frac{\bar{X}_1 - \bar{X}_2}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}
$

where $\bar{X}_1$ and $\bar{X}_2$ are the sample means, $s_1^2$ and $s_2^2$ are the sample variances, and $n_1$ and $n_2$ are the sample sizes.

**Calculation:**

1. **Calculate the sample means:**

   - Mean of Caucasian:
     $
     \bar{X}_1 = \frac{14 + 25 + 28 + 31 + 32 + 41 + 56 + 65 + 71 + 94}{10} = 45.7
     $

   - Mean of African American:
     $
     \bar{X}_2 = \frac{12 + 18 + 19 + 21 + 22 + 35 + 36 + 39 + 42 + 43 + 44 + 46 + 50 + 54 + 91}{15} = 37.5
     $

2. **Calculate the sample variances:**

   - Variance of Caucasian:
     $
     s_1^2 = \frac{\sum (X_i - \bar{X}_1)^2}{n_1 - 1} = \frac{(14-45.7)^2 + (25-45.7)^2 + \cdots + (94-45.7)^2}{10 - 1} = 636.4
     $

   - Variance of African American:
     $
     s_2^2 = \frac{\sum (X_i - \bar{X}_2)^2}{n_2 - 1} = \frac{(12-37.5)^2 + (18-37.5)^2 + \cdots + (91-37.5)^2}{15 - 1} = 504.9
     $

3. **Calculate the t-test statistic:**

   $
   t = \frac{45.7 - 37.5}{\sqrt{\frac{636.4}{10} + \frac{504.9}{15}}} = \frac{8.2}{\sqrt{63.64 + 33.66}} = \frac{8.2}{\sqrt{97.3}} = \frac{8.2}{9.86} = 0.8319
   $

4. **Determine the degrees of freedom:**

   Using the formula for Welch's degrees of freedom:
   $
   df = \frac{\left(\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}\right)^2}{\frac{\left(\frac{s_1^2}{n_1}\right)^2}{n_1 - 1} + \frac{\left(\frac{s_2^2}{n_2}\right)^2}{n_2 - 1}}
   $
   
   $
   df = \frac{\left(\frac{636.4}{10} + \frac{504.9}{15}\right)^2}{\frac{\left(\frac{636.4}{10}\right)^2}{9} + \frac{\left(\frac{504.9}{15}\right)^2}{14}} = \frac{\left(63.64 + 33.66\right)^2}{\frac{4047.68}{9} + \frac{2549.71}{14}} = 23.14
   $

5. **Compare the calculated t-test statistic with the critical value:**

   Using a t-distribution table, for df $\approx 23$ and $\alpha = 0.05$, the critical value is 2.069.

   Since $t_{calculated}$ = 0.8319 is less than $t_{critical}$ = 2.069$, we fail to reject the null hypothesis.

**Interpretation:** There is no significant difference in the distribution of serum estradiol between Caucasian and African American women at the 5% significance level.

$
P(X = 0) = e^{-\lambda}
$

where $\lambda$ is the expected number of events.

**Calculation:**

   - For Researcher I:
     $
     \lambda_I = \frac{20}{1000} \times 100 = 2
     $

   - For Researcher II:
     $
     \lambda_{II} = \frac{800}{80,000} \times 100 = 1
     $

2. **Calculate the probability:**

   - For Researcher I:
     $
     P(X_I = 0) = e^{-2} \approx 0.1353
     $

   - For Researcher II:

### Problem 2(a)

**Objective:** Test whether the true average stance duration is larger among elderly individuals than among younger individuals at a 0.05 significance level.

**Given Data:**

- **Older adults:**
  - Sample size ($n_1$): 28
  - Sample mean ($\bar{X}_1$): 801 ms
  - Sample standard deviation ($s_1$): 117 ms

- **Younger adults:**
  - Sample size ($n_2$): 16
  - Sample mean ($\bar{X}_2$): 780 ms
  - Sample standard deviation ($s_2$): 72 ms

**Formula:**

To compare the means, we use a two-sample t-test assuming unequal variances (Welch's t-test). The test statistic is:
$
t = \frac{\bar{X}_1 - \bar{X}_2}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}
$
The degrees of freedom for the test are calculated using:
$
df = \frac{\left(\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}\right)^2}{\frac{\left(\frac{s_1^2}{n_1}\right)^2}{n_1 - 1} + \frac{\left(\frac{s_2^2}{n_2}\right)^2}{n_2 - 1}}
$

**Calculation:**

1. **Calculate the test statistic:**

   $
   t = \frac{801 - 780}{\sqrt{\frac{117^2}{28} + \frac{72^2}{16}}} = \frac{21}{\sqrt{\frac{13689}{28} + \frac{5184}{16}}} = \frac{21}{\sqrt{489.6 + 324}} = \frac{21}{\sqrt{813.6}} = \frac{21}{28.53} = 0.736
   $

2. **Calculate the degrees of freedom:**

   $
   df = \frac{\left(\frac{117^2}{28} + \frac{72^2}{16}\right)^2}{\frac{\left(\frac{117^2}{28}\right)^2}{27} + \frac{\left(\frac{72^2}{16}\right)^2}{15}} = \frac{813.6^2}{\frac{239.57}{27} + \frac{202.5}{15}} = \frac{662044.96}{8.87 + 13.5} = 25.66
   $

3. **Determine the critical value:**

   For a one-tailed test at the 0.05 significance level and $df \approx 26$, the critical value from the t-distribution table is approximately 1.706.

4. **Compare the test statistic with the critical value:**

   Since $t_{calculated}$ = 0.736 is less than $t_{critical}$ = 1.706, we fail to reject the null hypothesis.

**Interpretation:** There is no significant evidence to conclude that the true average stance duration is larger among elderly individuals compared to younger individuals at the 5% significance level.

### Problem 2(b)(i)

**Objective:** Test whether the incidence rate of GI problems differs between the olestra treatment group and the TG control group at the 5% significance level.

**Given Data:**

- **TG control group:**
  - Number of individuals: 529
  - Number of GI events: 79

- **Olestra treatment group:**
  - Number of individuals: 563
  - Number of GI events: 113

**Formula:**

To compare the incidence rates, we use a two-proportion z-test. The test statistic is:
$
z = \frac{p_1 - p_2}{\sqrt{p(1-p)\left(\frac{1}{n_1} + \frac{1}{n_2}\right)}}
$
where:
$
p_1 = \frac{x_1}{n_1}, \quad p_2 = \frac{x_2}{n_2}, \quad p = \frac{x_1 + x_2}{n_1 + n_2}
$

**Calculation:**

1. **Calculate the proportions:**

   - Proportion in the TG control group:
     $
     p_1 = \frac{79}{529} = 0.1493
     $

   - Proportion in the olestra treatment group:
     $
     p_2 = \frac{113}{563} = 0.2007
     $

   - Combined proportion:
     $
     p = \frac{79 + 113}{529 + 563} = \frac{192}{1092} = 0.1758
     $

2. **Calculate the z-test statistic:**

   $
   z = \frac{0.1493 - 0.2007}{\sqrt{0.1758(1-0.1758)\left(\frac{1}{529} + \frac{1}{563}\right)}} = \frac{-0.0514}{\sqrt{0.1758 \times 0.8242 \times \left(0.00189 + 0.00178\right)}} = \frac{-0.0514}{\sqrt{0.1758 \times 0.8242 \times 0.00367}} = \frac{-0.0514}{0.0264} = -1.947
   $

3. **Determine the critical value:**

   For a two-tailed test at the 0.05 significance level, the critical value is approximately 1.96.

4. **Compare the test statistic with the critical value:**

   Since $|z_{calculated}|$ = 1.947 is less than $z_{critical}$ = 1.96, we fail to reject the null hypothesis.

**Interpretation:** There is no significant evidence to conclude that the incidence rate of GI problems differs between the olestra treatment group and the TG control group at the 5% significance level.

### Problem 2(b)(ii)

**Objective:** Design a new study with 90% power to detect a difference, with four times the number of subjects in the control group as in the treatment group.

**Given Data:**

- **Previous proportions:**
  - $p_1$ = 0.1493 (TG control group)
  - $p_2$ = 0.2007 (Olestra treatment group)

**Formula:**

To calculate the required sample size for each group, we use the formula for sample size in a two-proportion z-test:
$
n_2 = \left(\frac{Z_{\alpha/2} + Z_{\beta}}{p_1 - p_2}\right)^2 \times p(1-p)
$
where:
- $Z_{\alpha/2}$ is the z-value for the significance level (1.96 for 0.05).
- $Z_{\beta}$ is the z-value for the power (1.28 for 90% power).
- $p = \frac{p_1 + p_2}{2}$ is the average proportion.

The sample size for the control group $n_1$ is four times the sample size for the treatment group.

**Calculation:**

1. **Calculate the average proportion:**

   $
   p = \frac{0.1493 + 0.2007}{2} = 0.175
   $

2. **Calculate the required sample size for the treatment group:**

   $
   n_2 = \left(\frac{1.96 + 1.28}{0.2007 - 0.1493}\right)^2 \times 0.175 \times (1-0.175)
   $
   $
   n_2 = \left(\frac{3.24}{0.0514}\right)^2 \times 0.175 \times 0.825
   $
   $
   n_2 = \left(63.08\right)^2 \times 0.1444 = 3974.35 \times 0.1444 \approx 573.89
   $

   Therefore, $n_2$ $\approx 574$ for the treatment group.

3. **Calculate the required sample size for the control group:**

   $
   n_1 = 4 \times n_2 = 4 \times 574 = 2296
   $

**Interpretation:** To achieve 90% power to detect a difference in the incidence rate of GI problems between the olestra treatment group and the TG control group, the new study should include approximately 574 subjects in the treatment group and 2296 subjects in the control group.
